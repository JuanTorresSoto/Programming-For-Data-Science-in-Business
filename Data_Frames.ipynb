There are two main data types in pandas

1. series it is a list containing data of any one data type. think of it like a vector from R or list from python

2. Dataframe - a table with rows and columns like a dataframe in R or a table in Excel

```python
s = pd.Series(["a","b","c","d"])
s
```

```python
sdata= {"a": 100, "b": 200, "c": 300, "d": 400}
s = pd.Series(sdata)
s
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories, index =["day1","day2"])
print(myvar)
```

**We can make pandas data frames in two ways**

1. With a dictionary where the key will be the column name and value list will be row entries

2. With a nested dictionary where we have the structure{colname:value,rowname:value},colname2:{rowname:value,rowname:value,rowname:value}}

3. Zip multiple list together to create tuples and then convert that top a dataframe

```python
names = ["Bob","Jessica","Mary","John","Mel"]
births = [968, 155, 77, 578, 973]

babyDataSet = list(zip(names,births))
babyDataSet
```

```python
df = pd.DataFrame(data = babyDataSet)
df
```
