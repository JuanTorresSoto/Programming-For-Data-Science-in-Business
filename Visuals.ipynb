There are two types of viz:

* Univariate - one variable - purpose to understand the purpose of that variable

1. Pie chart
2. Bar charts
  -cluster bar chart
  -stacked bar chart
3. Box plot
4. Histograms: is a frequency distribution of a numeric variables
5. Violin plot

* Multivariate - You can usually do at most 4 variables in a visualization

1. Scatter plot: two continuous variables
2. Line plot: usually one date axis and one numeric variable
3. Title chart and Bubble chart: two numeric variables for x and y axis, one numeric variable for size of title/bubble and one categorical variable for cololr of size/bubble


```python
import seaborn as sns
import matplotlib.pyplot as plt

#load data
df = sns.load_dataset('iris')
df
```

```python
#displot function creates a histogram
#bins refers to intervals or ranges used to group data
sns.displot(df["sepal_length"], bins = 10)
```

```python
#kde stands for kernel density
sns.displot(df["sepal_length"], bins = 10, kde = True)
plt.show()
```

```python
sns.displot(df["sepal_length"], bins = 10, stat = "density")
plt.show()
```

```python
sns.kdeplot(df["sepal_length"])
plt.show()
```

```python
sns.kdeplot(df["sepal_length"], fill = True, color = "red")
plt.show()
```

```python
sns.histplot(df["sepal_width"], label = 'sepal_width', color = 'crimson', stat = "density")
sns.histplot(df["sepal_length"], label = 'sepal_length', color = 'gold', stat = "density")
plt.legend()
plt.show()
```

```python
sns.histplot(df["sepal_width"], label = 'sepal_width', color = 'darkgreen', kde = True, stat = "density")
sns.histplot(df["sepal_length"], label = 'sepal_length', color = 'gold', kde = True,  stat = "density")
plt.legend()
plt.show()
```

```python
sns.kdeplot(x = df['sepal_length'], y = df['sepal_width'], fill = True, color = "blue")
```

```python
#using subplot to create multiple histograms
#plt.subplots -- the first two arguments tell number of rows and columns. figsize tells size of the charts
#sharex means that values will share the x axis
f, axes = plt.subplots(2,2, figsize=(7,7), sharex = True)
sns.histplot(df['sepal_length'], color ="lightgreen", ax = axes[0,0])
sns.histplot(df['sepal_width'], color ="red", ax = axes[0,1])
sns.histplot(df['petal_length'], color ="blue", ax = axes[1,0])
sns.histplot(df['petal_width'], color ="gold", ax = axes[1,1])
```

**NOTES**

kind = scatter will give you a scatter plot and univariate histograms

kind - "kde" will give you contour plot and density plots

kind = "hex" will give you hexagonal titles and density plots

kind = "hist" will give title plot and histograms

kind = "reg" gives you a regression line

```python
sns.jointplot(x = df['sepal_length'], y = df['sepal_width'], kind='scatter')
```

```python
sns.jointplot(x = df['sepal_length'], y = df['sepal_width'], color = "orange", kind='kde')
```

```python
sns.jointplot(x = df['sepal_length'], y = df['sepal_width'], color = "red", kind='hex')
```

```python
sns.jointplot(x = df['sepal_length'], y = df['sepal_width'], color = "darkgreen", kind='hist')
```

```python
sns.jointplot(data = df, x = 'sepal_length', y = 'sepal_width', hue = "species")
```

```python
sns.regplot(x = df['sepal_length'], y = df['sepal_width'])
```

```python
sns.lmplot(data = df, x = 'sepal_length', y = 'sepal_width', hue = "species", fit_reg= True,
           legend = True, markers =["o","x","1"])
```

```python
custom_pal={'setosa':'lightgreen','versicolor':'olive','virginica':'darkgreen'}
sns.boxplot(x = df['species'], y = df['sepal_length'], palette = custom_pal)
```
